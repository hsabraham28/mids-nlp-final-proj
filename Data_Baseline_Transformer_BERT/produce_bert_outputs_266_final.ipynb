```python
import pandas as pd
# import keras_nlp
import tensorflow as tf
from tensorflow import keras
import numpy as np
import random
from transformers import BertTokenizer, TFBertModel
```

## 5k dataset

```python
com_df = pd.read_csv('/content/drive/MyDrive/266_data/combined_df.csv', usecols=['TI', 'EN', 'TI_tokenized'])
```

```python
bi_dat_2 = pd.read_csv('/content/drive/MyDrive/266_data/bi_dat_2.csv', usecols = ['TI', 'EN', 'TI_tokenized', 'EN_tokenized'])
```

```python
# eng_list = bi_dat_2['EN'].values.tolist()
eng_list = com_df['EN'].values.tolist()
```

```python
eng_list_1000_1 = eng_list[3000:]
```

```python
bert_tokenizer = BertTokenizer.from_pretrained('bert-base-cased')
```

```python
tokenized_input = bert_tokenizer(eng_list,
                                 max_length=30,
                                 truncation=True,
                                 padding='max_length',
                                 return_tensors='tf')
```

```python
bert_model = TFBertModel.from_pretrained('bert-base-cased')
```

```python
outputs = bert_model(tokenized_input, output_hidden_states=True)
```

```python
bert_embedding = outputs.hidden_states[0].numpy()
```
