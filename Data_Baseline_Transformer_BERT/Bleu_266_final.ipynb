## Results

```python
results_ts = pd.read_csv('/content/drive/MyDrive/266_data/bert_mod_ts_8_4.csv', usecols=['actual', 'predicted'])
```

```python
ref_list = results_ts['actual'].values.tolist()
```

```python
pred_list = []
# test_list = []

# idx_test = [1, 2, 3]
# for i in idx_test:
for i in results_ts.index:
  end_idx = results_ts['predicted'].loc[i].find('[END]')
  if end_idx == -1:
    # print(i)
    end_idx = results_ts['predicted'].loc[i].find('[PAD]')
  clean_str = results_ts['predicted'].loc[i][:end_idx]
  # print(clean_str)
  pred_list.append(clean_str)
```

```python
bleu = evaluate.load('bleu')

results = bleu.compute(predictions=pred_list,
                        references=ref_list,
                       max_order=4) # set n-gram
print(results)
```
